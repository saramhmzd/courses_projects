```python
import matplotlib.pyplot as plt
import numpy as np


images = ['1', '2', '3']
for i in images:
  img = plt.imread(i+'.bmp')
  print("-------------------------------------------------")
  plt.imshow(img)
  plt.show()
  print("-----------------------")
  print("Shape of the loaded image is:", img.shape)
  print("-----------------------")
  print("Image as numpy array:")
  print(img)
  print("-------------------------------------------------")

  r = img[:, :, 0]
  b = img[:, :, 1]
  g = img[:, :, 2]

  u_r, s_r, v_r = np.linalg.svd(r)
  u_b, s_b, v_b = np.linalg.svd(b)
  u_g, s_g, v_g = np.linalg.svd(g)
  
  for k in [10 , 50, 150, 250]:
    print('----------------------------------')
    print('----------------------------------')
    r_k = np.dot(u_r[:, :k],np.dot(np.diag(s_r[:k]), v_r[:k, :]))
    b_k = np.dot(u_b[:, :k],np.dot(np.diag(s_b[:k]), v_b[:k, :]))
    g_k = np.dot(u_g[:, :k],np.dot(np.diag(s_g[:k]), v_g[:k, :]))
    A_k = np.zeros((1280,1920,3), 'uint8')
    A_k[..., 0] = r_k
    A_k[..., 1] = b_k
    A_k[..., 2] = g_k
    print('k = ', k)
    image = plt.imshow(A_k)
    plt.show()
```
